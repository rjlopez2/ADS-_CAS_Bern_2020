```python
def get_merged_csv(flist, **kwargs):
    file_path = os.path.join("WB_Geo_and_Eco_groups.csv")
    geo = pd.read_csv(file_path)
    for file in flist:
        my_indi = pd.read_csv(file, **kwargs)
        geo = geo.merge(my_indi, how= "left")
    return(geo)
```
